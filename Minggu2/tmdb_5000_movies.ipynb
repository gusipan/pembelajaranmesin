```python
import ast
df['production'] = df['production_countries'].apply(ast.literal_eval)

# Extract 'id' and 'name' separately
df['production_country_codes'] = df['production'].apply(lambda x: [d['iso_3166_1'] for d in x])
df['production_country'] = df['production'].apply(lambda x: [d['name'] for d in x])
df.drop(columns=['production'], inplace=True)
df.drop(columns=['production_countries'], inplace=True)
```

```python
import ast
df['lan'] = df['spoken_languages'].apply(ast.literal_eval)

# Extract 'id' and 'name' separately
df['spoken_language_codes'] = df['lan'].apply(lambda x: [d['iso_639_1'] for d in x])
df['spoken_languages_'] = df['lan'].apply(lambda x: [d['name'] for d in x])
df.drop(columns=['lan'], inplace=True)
df.drop(columns=['spoken_languages'], inplace=True)
```

```python
import ast
df['company'] = df['production_companies'].apply(ast.literal_eval)

# Extract 'id' and 'name' separately
df['production_compony_ids'] = df['company'].apply(lambda x: [d['id'] for d in x])
df['production_companies_'] = df['company'].apply(lambda x: [d['name'] for d in x])
df.drop(columns=['company'], inplace=True)
df.drop(columns=['production_companies'], inplace=True)
```

```python
def search_trends_by_year(year):
    filtered_df = df[df['release_date'].dt.year == year]
    filtered_df["overall_ranking"]=filtered_df["vote_average"]*filtered_df["popularity"]
    best_voted= filtered_df.sort_values(by='vote_average', ascending=False)
    most_popular= filtered_df.sort_values(by='popularity', ascending=False)
    overall= filtered_df.sort_values(by='overall_ranking', ascending=False)
    return overall[["id","original_title", "vote_average", "popularity", "overall_ranking", "release_date"]].head(10)
```

```python
search_trends_by_year(2005)
```
